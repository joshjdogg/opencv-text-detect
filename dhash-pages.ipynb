```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set()
import dhash
from imutils import paths
from PIL import Image
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
def convertNumberToArray(number, size=8):
    output = '{:0{}b}'.format(number, size * size)
    output
    output = np.array([[int(output[i+x])for x in range(size)] for i in range(0, size * size, size)])
    return output
```

```python
def showhash(hash, desc, refHash=None, size=8):
    hashArray = convertNumberToArray(hash, size=size)
    newhash = np.asarray(dtype=np.dtype('uint8'), a=[[0 if i == 0 else 225 for i in x] for x in hashArray])
    hashimg = Image.fromarray(newhash, mode='L')
    refimg = Image.new('L', (0,0), (255))
    hash
    print(desc)
    if refHash is None:
        sns.heatmap(hashArray, cbar=False, robust=True, square=True)
    else:
        refhashArray = convertNumberToArray(refHash, size=size)
        newRefhash = np.asarray(dtype=np.dtype('uint8'), a=[[0 if i == 0 else 225 for i in x] for x in refhashArray]) 
        refimg = Image.fromarray(newRefhash, mode='L')
        refHash
        fig, ax = plt.subplots(1,2)
        sns.heatmap(hashArray, cbar=False, robust=True, square=True, ax=ax[0])
        sns.heatmap(refhashArray, cbar=False, robust=True, square=True, ax=ax[1])
    plt.show()

    return hashimg, refimg
   
    
```

```python
def diffimg(i1, i2):
    #i1 = Image.open("images/car_wash.png")
    #i2 = Image.open("images/car_wash_mod.png")
    assert i1.mode == i2.mode, "Different kinds of images."
    assert i1.size == i2.size, "Different sizes."

    pairs = zip(i1.getdata(), i2.getdata())
    if len(i1.getbands()) == 1:
        # for gray-scale jpegs
        dif = sum(abs(p1-p2) for p1,p2 in pairs)
    else:
        dif = sum(abs(c1-c2) for p1,p2 in pairs for c1,c2 in zip(p1,p2))

    ncomponents = i1.size[0] * i1.size[1] * 3
    diffpercent = (dif / 255.0 * 100) / ncomponents
    print ("Difference (percentage):", diffpercent, "%")
    return diffpercent
```
